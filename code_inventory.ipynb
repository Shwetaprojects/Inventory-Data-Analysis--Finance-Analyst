```python

plt.figure(figsize=(3,3))
mylabels=["MARTIGNETTI COMPANIES","ULTRA BEVERAGE COMPANY LLP",'M S WALKER INC','PERFECTA WINES','E & J GALLO WINERY']
plt.pie(Purchase_PricesDec.VendorName.value_counts()[0:5],labels=mylabels ,autopct='%0.1f%%',radius=2,wedgeprops={'edgecolor':'white'},textprops={'size':9, },shadow=True)
plt.title('Top 5 vendors\n')
plt.show()
```

```python
plt.figure(figsize=(15,5))

vc = Purchase_PricesDec.groupby(['Description'])['Price'].max().sort_values(ascending=False)[:10]


plt.figure(figsize=(12, 6)) 
g = sns.barplot(x=vc.index, y=vc.values)


for i, value in enumerate(vc):
 g.text(x=i, y=value + 0.5,s=f'{value:.2f}', color='black', ha='center', fontsize=10)


plt.title('Top 10 Brands with Highest Price Tags')
plt.xlabel('Brands')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.tight_layout() 
plt.show()
```
